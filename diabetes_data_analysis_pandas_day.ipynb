## Filters
1. BMI over 35

```python
high_bmi = df[df['BMI'] > 35]
print(high_bmi)
print("High BMI count:", len(high_bmi))
```

2. Patients above 60 with diabetes.

```python
older_diabetics = df[(df['Age'] > 60) & (df['Outcome'] == 1)]
print(older_diabetics)
print("Older and Diabetic count:", len(older_diabetics))
```

3. Low Glucose but High Insulin

```python
odd_insulin = df[(df['Glucose'] < 100) & (df['Insulin'] > 110)]
print(odd_insulin)
print("Odd Insulin count:", len(odd_insulin))
odd_insulin.plot(kind="scatter", x="Glucose", y="Insulin", title="Low Glucose, High Insulin")
```

## Ploting

```python
df["Age"].plot(kind="hist", bins=25)
plt.title("Age Histogram")
plt.show()

df.plot(kind="scatter", x="BMI", y="Glucose")
plt.title("BMI vs Glucose")
plt.show()


df["BMI"].plot(kind="hist", bins=20)
plt.title("BMI Histogram")
plt.show()

df["Pregnancies"].plot(kind="hist", bins=1)
plt.title("Pregnancies Histogram")
plt.show()

df.plot(kind="scatter", x="BMI", y="Glucose")
plt.title("Outcome vs Glucose")
plt.show()
```

## Risk Classifier

```python
def risk_classify(row):
  if row["BMI"] > 35 and row["Glucose"] > 140:
        return "High Risk"
  else:
        return "Low Risk"

df["Risk_Level"] = df.apply(risk_classify, axis=1)
df["Risk_Level"].value_counts().plot(kind="bar")
plt.title("Patient Risk Levels")
plt.show()
```

## Flag Patients with Possible Pre-Diabetes

```python
def possible_pre_diabetes(row):
  if 100 <= row["Glucose"] <= 120 & row["Outcome"] == 0:
    return "Possible Diabetes"
  else:
     return "No Possble diabeties"

df["Pre Diabetic Flag"] = df.apply(possible_pre_diabetes, axis=1)
print(df[["Glucose", "Outcome", "Pre Diabetic Flag"]])
```
